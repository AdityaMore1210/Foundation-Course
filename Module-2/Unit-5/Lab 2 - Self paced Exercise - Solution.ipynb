## Count Plot:
Create a count plot to visualize the distribution of flights across months.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Create a count plot
plt.figure(figsize=(10, 6))
sns.countplot(x="month", data=flights, palette="viridis")
plt.title("Distribution of Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

### Smart Insights: Countplot

- The count plot shows that the highest number of flights occurred in the month of August, while the lowest number was observed in February.

## Strip Plot:
Generate a strip plot to visualize the distribution of passenger counts across months.

```python
# Create a strip plot
plt.figure(figsize=(10, 6))
sns.stripplot(x="month", y="passengers", data=flights, jitter=True, palette="plasma")
plt.title("Passenger Count Distribution by Month (Strip Plot)")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
plt.show()
```

### Smart Insights: Strip Plot

- The strip plot reveals the distribution of passenger counts across different months. It indicates higher passenger counts during the summer months, particularly from May to August.

## Swarm Plot:
Create a swarm plot to better understand the distribution of passenger counts across months.

```python
# Create a swarm plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x="month", y="passengers", data=flights, palette="coolwarm")
plt.title("Passenger Count Distribution by Month (Swarm Plot)")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
plt.show()
```

### Smart Insights: Swarm Plot

- The swarm plot provides a detailed view of passenger count distribution. It further highlights the concentration of passengers during peak travel months.
