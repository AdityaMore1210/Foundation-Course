# Lab-1 - Self paced Exercise - Solution : Analyzing Flights Dataset with Seaborn

# Dataset:
Seaborn's built-in "flights" dataset.

# Objective: 

Explore the monthly passenger trends, visualize the correlation between years and months, and observe the changes in passenger numbers over the years.

# Steps:

## Import necessary libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")
```

## Load the dataset

## Scatter Plot:

Create a scatter plot to visualize the monthly passenger trends over the years.

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="year", y="passengers", data=flights, hue="month", palette="viridis")
plt.title("Monthly Passenger Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend(title="Month")
plt.show()
```

### Smart Insights: Scatter Plot (Monthly Passenger Trends Over Years):

- The scatter plot illustrates the monthly passenger trends over the years.
- There is a general upward trend in the number of passengers over the years.
- Some months show higher passenger counts, indicating potential seasonal variations.

## Heatmap:
Generate a heatmap to show the correlation between years and months.

```python
# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Passenger Count by Year and Month")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

### Smart Insights: Heatmap (Passenger Count by Year and Month)


- The heatmap displays the correlation between years and months in terms of passenger counts.
- The color intensity increases for years with higher passenger counts during certain months, indicating peak travel periods.
- The passenger count tends to increase over the years for specific months.

## Line Plot:
Plot the trend of passenger numbers over the years.

```python
# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="year", y="passengers", data=flights, ci=None)
plt.title("Passenger Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()
```

### Smart Insights: Line Plot (Passenger Trends Over Years)

- The line plot depicts the trend of passenger numbers over the years.
- There is a consistent overall increase in the number of passengers from year to year.
- There is a noticeable upward trend with slight fluctuations.
