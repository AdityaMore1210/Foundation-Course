# Lab-3 - Self paced Exercise - Solution : Analyzing Flights Dataset with Seaborn

# Dataset:
Seaborn's built-in "flights" dataset.

# Objective: 

Visualize the distribution of continuous numerical data.

# Steps:

## Import necessary libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Load the dataset

```python
# Load the flights dataset
flights = sns.load_dataset("flights")
```

## Instructions:
Writedown the pointswise insights below all the charts respectively.

## Histograms:

Create histograms to visualize the distribution of passenger counts.

```python
# Create histograms
plt.figure(figsize=(10, 6))
sns.histplot(data=flights, x="passengers", bins=20, kde=True, color="blue")
plt.title("Passenger Count Distribution (Histogram)")
plt.xlabel("Number of Passengers")
plt.ylabel("Frequency")
plt.show()
```

### Smart Insights:Histograms

- The histograms depict the distribution of passenger counts. They show that the most common range of passenger counts is around 200-300, followed by 400-500.

## KDE (Kernel Density Estimate) Plot:
Generate KDE plots to observe the density estimation of passenger counts.

```python
# Create KDE plots
plt.figure(figsize=(10, 6))
sns.kdeplot(data=flights, x="passengers", hue="month", fill=True, common_norm=False)
plt.title("Passenger Count Distribution by Month (KDE)")
plt.xlabel("Number of Passengers")
plt.ylabel("Density")
plt.legend(title="Month")
plt.show()
```

### Smart Insights:KDE Plot

- The KDE plot illustrates the density estimation of passenger counts across different months. It highlights the fluctuations in density, indicating peak travel months with higher passenger counts.
